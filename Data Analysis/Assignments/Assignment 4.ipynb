```python
import numpy as np
import pandas as pd
```

### 1. Create a DataFrame:
o Load the above dataset into a Pandas DataFram

```python
sales = pd.DataFrame({'Order ID':[1001,1002,1003,1004,1005,1006,1007,1008],
                      'Customer':['Alice','Bot','Charlie','Alice','David','Charlie','Bob','Alice'],
                      'Product':['Laptop','Phone','Chair','Table','Laptop','Phone','Chair','Phone'],
                      'Category':['Electronics','Electronics','Electronics','Electronics','Electronics','Electronics','Electronics','Electronics'],
                      'Quantity':[2,1,4,1,1,2,2,1],
                      'Price':[800,500,100,250,800,500,100,500],
                      'Total Sales':[1600,500,400,250,800,1000,200,500]})
sales
```

### 2. Add a New Column:
o Add a column Discount that gives a 10% discount on total
sales.


```python
sales['Discount']= sales['Total Sales']*0.10
```

```python
sales
```

### 2. Add a New Column:
o Add a column Discount that gives a 10% discount on total
sales.

```python
new_row=pd.DataFrame([{'Order ID':1009, 
                       'Customer':'Eve',
                        'Product':'Tablet',
                        'Category':'Electronics',
                        'Quantity':1,
                        'Price':300,
                        'Total Sales':300}])
```

```python
new_row['Discount']=new_row['Total Sales']* 0.10
```

```python
sales=pd.concat([sales, new_row])
```

```python
sales
```

```python
sales.reset_index(inplace=True)
```

```python
sales
```

```python
sales.drop('index',axis=1, inplace=True)
```

```python
sales
```

### 4. Filter Data:
o Retrieve only the sales records where the Category is
"Electronics".

```python
electronic_sales=sales[sales['Category']=='Electronics']
```

```python
electronic_sales
```

### 5. Selection:
o Select the Customer and Total Sales columns.

```python
customer_sales=sales[['Customer', 'Total Sales']]
```

```python
customer_sales
```

### 6. Grouping:
o Group the data by Customer and calculate the total sales per
customer.

```python
total_sales= sales.groupby('Customer')['Total Sales'].sum()
```

```python
total_sales
```

### 7. Sorting:
o Sort the DataFrame based on Total Sales in descending
order.

```python
sales_sorted= sales.sort_values(by='Total Sales',ascending=False)
```

```python
sales_sorted
```

### Apply Function:
o Apply a lambda function to create a new column Sales
Category, where values are labeled as "High" if Total Sales >
1000, otherwise "Low"

```python
sales['Sales Category']= sales['Total Sales'].apply(lambda x:'High' if x>1000 else 'Low')
```

```python
sales
```

### 9. Save DataFrame:
o Save the final DataFrame as a CSV file named sales_data.csv.

```python
sales.to_csv('sales_data.csv', index=False)
```
