```python
import matplotlib.pyplot as plt
import seaborn as sns
```

### Bar chart- Department wise average performance

```python
avg=data.groupby('Department')['Performance_Score'].mean()
plt.bar(avg.index, avg.values)
plt.title('Average Performance by Departmnet')
plt.xlabel('Department')
plt.ylabel('Performance Score')
plt.show()
```

### Boxplot- Salary distribution by department

```python
data.boxplot(column='Salary(LPA)',by='Department')
plt.title('Salary Distribution by department')
plt.xlabel('Department')
plt.ylabel('Salary(LPA)')
plt.show()
```

### Histogram- Age disribution of employees

```python
plt.hist(data['Age'],edgecolor='black',bins=10)
plt.title('Age Distribution of Employees')
plt.xlabel('Age')
plt.ylabel('Number of Employee')
plt.show()
```

### Heatmap- Correlation matrix
