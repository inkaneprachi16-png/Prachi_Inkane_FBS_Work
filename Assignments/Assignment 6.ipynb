```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
car=pd.read_csv('Datasets/cars93.csv')
```

```python
car
```

### 1. Bar Chart:
a. Create a bar chart to compare the average price of cars by manufacturer.
b. Label the axes and add a suitable title.
c. Use different colors for different manufacturers.

```python
avg_price= car.groupby('Manufacturer')['Price'].mean()
avg_price.plot.bar(color=plt.cm.tab20.colors)
plt.xlabel('Manufacturer')
plt.ylabel('Average Price')
plt.title('Average car price by manufacturer')

```

### Line Chart:
a. Create a line chart showing the trend of horsepower for different types of cars.
b. Include proper legends, labels, and a title

```python
line_chart=car.groupby('Type')['Horsepower'].mean()
```

```python
line_chart.plot(kind='line', marker='o')
plt.xlabel('Car Type')
plt.ylabel('Average Horsepower')
plt.title('Trend of Horsepower by car types')
plt.grid(True)
plt.show()
```

### Pie Chart:
a. Create a pie chart showing the proportion of different car types in the dataset.
b. Include percentages in the chart

```python
type_count= car['Type'].value_counts()
```

```python
type_count
```

```python
values=type_count.values
labels=type_count.index
plt.pie(values, labels=labels, autopct='%0.2f%%')
plt.title('Proportion of Different car types')
plt.show()
```

### Scatter Plot:
a. Create a scatter plot to show the relationship between horsepower and MPG
(Miles Per Gallon).
b. Label the axes and add a title.

```python
plt.scatter(car['Horsepower'], car['MPG.city'], color='red')
plt.xlabel('Horesepower')
plt.ylabel('Miles Per Gallon(city)')
plt.title('Relationship Between Horesepower and MPG (city)')
```
