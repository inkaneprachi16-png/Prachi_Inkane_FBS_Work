```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = pd.read_csv('Datasets/tips.csv')
```

```python
tips
```

### 1. Plot a day wise boxplot of totbill column

```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.show()
```

### 2. plot a pariplot by setting hue parameter t

```python
sns.pairplot(tips, hue='time')
```

### 3. Draw a scatter plot of tip and totalbill and also allot the points according to the sex


```python
sns.scatterplot(x='total_bill',y='tip',hue='sex',data=tips)
```

### 4. Plot a hitsogram of totbill for 50 bins


```python
plt.hist(tips['total_bill'], edgecolor='black', bins=10)
plt.show()
```

### 5. Plot a heatmap to represent entire dataset's correlation

```python
tips.info()
```

```python
correlation_tips=tips[['total_bill','tip','size']]
```

```python
sns.heatmap(correlation_tips.corr(),annot=True)
```
